# Hyperbolic Signal-Response System

```python
from ulbc import *
from flowstar import *
from ulbc.bondcalculus import *
from scipy.integrate import solve_ivp
from ulbc.matricies import vec_to_numpy, mat_to_numpy, convert_vec, sympy_gens

import numpy as np
from matplotlib import pyplot as plt

from contextlib import contextmanager
import sys, os

@contextmanager
def suppress_stdout():
    with open(os.devnull, "w") as devnull:
        old_stdout = sys.stdout
        sys.stdout = devnull
        try:  
            yield
        finally:
            sys.stdout = old_stdout
```

## Load the bond-calculus system

```python
# Replace the path with your example code if you want
m = BondModel("/lbuc/hostnotebooks/models/elementary/B_P_DeP_MA.bond")
p = m.process("Pi")
s = p.as_system; s
```

### Simple plotting

```python
r = s.reach(20,step=0.001)
r.successful
```

```python
plot((lambda t: r(t)[0].center(),
      lambda t: r(t)[2].center()),
      (0,20))
```

```python
s1 = s.with_y0([0,2,1])
r = s1.reach(20, step=0.001)
r.successful
```

```python
plot((lambda t: r(t)[0].center(),
      lambda t: r(t)[2].center()),
      (0,20))
```

### Repeat with uncertain initial value for S

```python
s2 = s.with_y0([0,RIF(2,2.5),1])
r = s2.reach(10, step=0.001)
r.successful
```

```python
plot((lambda t: r(t)[0].upper(),
      lambda t: r(t)[0].center(),
      lambda t: r(t)[0].lower()),
      (0,10))
```

### Generate signal-response curve

```python
# Iterate through different initial values of S
sols = []
rng = np.arange(0,10.1,0.5)
with suppress_stdout():
    for signal in rng:
        temp_sys = s.with_y0([0,signal,1])
        sols.append(temp_sys.reach(10, step=0.001)(10))
sols = np.array(sols)
plt.plot(rng,sols[:,0])
```
